# Python revision

### A. Mutable data types
Can be altered & re-assigned in place. Lists, dicts , sets

```python
# Starting with list
alist = [11,22,33,44,55,66,77,88]
print(type(alist))

# calling by index
print(alist[-1])

# slicing
print(alist[1:4])

# printing in reverse order
print(alist[::-1])

# re-assigning values at an index
alist[2] = 100
print(alist)
```

```python
# Moving to dictionaries
d = {"a":11 , "b":22 , "c":33 , "d":44 , "d":55 , "e":66}
print(type(d))

# accessing a particular key-value pair
# convert d.items() into a list
print(list(d.items()))
# and then call by index
print(list(d.items())[2])
print(list(d.items())[1:4])

# re-assigning values
d["a"] = 100
print(d)

```

```python
# moving into sets - it accepts each value only once.
s = {1,3,4,5,7,7}
print(len(s))

# convert/ re-assign to a list to access the element by indexing
s_to_list = list(s)
print(type(s_to_list))
print(s_to_list[2]) 

# thus value can be re-assigned in a list & converted back to a set
```

### B. Mutable data types
CANNOT be altered nor re-assigned in place. Strings, tuples, boolean

```python
t = (11,2,33,44)
print(type(t))
# elements an be accessed by indexes
print(t[-1])
```

```python
# values can be re-assigned by converting the tuple to a dictionary
newlist = list(t)
print(newlist)
newlist[-1] = 55
print(newlist)
t = tuple(newlist)
print(t)
print(type(t))
```

### Re-assignment vs mutation
Its all about whether a variable is pointing to an object(list in this case) OR altering the object itself

```python
# This is reassignment
aa = alist + [99]
print(f"aa points to a new list-> {aa}")
print(f"the original list remains same-> {alist}")
```

```python
# This is mutation
ab = aa.append(99)
print(f"the original list aa itself is now altered->{aa}")
print(ab)
```

## Iterations & IO operations

```python
# iterating using simple 'for' loop -- just for accessing values
# Printing all the elements in the list
for  val in alist:
    print(f"val:{val}")

print("\n")

# range() enables access to index as well as values of the list
for  i in range(len(alist)):
    print(f"index-{i} val-{alist[i]}")
```

### Parameters vs Arguments distinction
in `adding_two_nos(val1, val2)` - `val1` `val2` are **parameters** -- these are part of the function's signature

The actual integer values that user will pass while running the function will be the **arguments**

```python
# Another way of writing the function
def adding_two_nos(val1, val2):
    """
    Desc: 
    ---------
    Adds the two integers entered by the user after validating the input values. 
    Throws an error in case of  invalid inputs.
    
    Args:
    ---------
    value 1: the first value to be added
    value 2: the second value to be added 
    
    Returns:
    ---------
    A string with the result or error
    """
    try:
        user_input1_val=int(val1)
        user_input2_val=int(val2)
        result=user_input1_val + user_input2_val

    
    except ValueError:
        return print("Enter valid input")

    return print(f"Result: {val1} + {val2} = {result}")
```

### Functions follow the LEGB rule for accessing a variable's scope : 
Variables are accessed in order of their availability starting from:<br> 
**Local --> Enclosing --> Global --> Built-in**  

```python
x = 50

def one():
  x = 10

def two():
  global x
  x = 30

def three():
  x = 100
  print(x)

for func in [one, two, three]:
  func()
  print(x)
```

### How are Performance score & Salary correlated?
Applying a logical order to the performance score using the `pd.CategoricalDtype`. So **Poor performance < Average < Good < Excellent**.

And to implement this, creating a new column that map the performance scores to a numerical value (1,2,3,4).

Since 

```python
perf_order = ["Poor", "Average", "Good", "Excellent"]
# Setting the ordered param to true
category_tmp = pd.CategoricalDtype(categories=perf_order,ordered=True)
df_employees["Performance_Score"] = df_employees["Performance_Score"].astype(category_tmp)
```

```python
# Mapping the scores to their respective value in na new column
df_employees["Performance_index"] = df_employees["Performance_Score"].map({"Poor":1, "Average":2, "Good":3, "Excellent":4})
df_employees["Performance_index"]= df_employees["Performance_index"].astype("int")
df_employees.head()
```

```python
# A boxplot gives the 5-number summary(mean,quartiles & range) for salaries for every performance score
# This gives a good understanding of how the salaries are distributed & compare them amongst categories
sns.boxplot(data=df_employees, x="Performance_Score", y="Salary", orient="v", hue="Performance_Score")
plt.suptitle("Salary distribution across Performance groups")
plt.title("Performance shows little correlation to Pay", fontstyle="italic", color="gray")

plt.show()
```

>Analyzing the boxplot:
> - Medians have identical levels across all groups. If salaries & performance were strongly correlated, then increasing levels of medians would have been observed (from poor rating to Excellent).
> - Overall height of the boxplots largely remains same across performance categories - this means the spread of salaries (IQR) is similar across all performance groups.
> - Box whiskers also have similar levels across performance groups. This indicates even the extreme minimum & maximum salaries are identical across the groups.

```python
# Plotting the correlation using spearman correlation
sns.heatmap(df_employees[["Salary","Performance_index"]].corr(numeric_only=True, method="spearman"),annot=True)

plt.suptitle("Correlation amongst the quantitative variables")
plt.title("Performance shows little correlation to Pay", fontstyle="italic", color="gray")
plt.show()
```

### Using pivot tables for aggregation
`df.pivot_tables()` creates a 'wide', cross-tabulated form of the dataframe.  .Important params of the method are:

`index` - column by which to group the dataframe , to use as row labels in the new table.

`columns` -  column(s) to use as column headers in the new table.

`values` - columns to be aggregated, either pass a list of columns or else all the numerical columns are aggregated.

`aggfunc` - aggregation function to be use (default is mean)

`fill_value` - default value ot be filled for NaN entries after aggregation

```python
# Find max salary per department
df_employees.pivot_table(index="Dept", values="Salary", observed=False, aggfunc="max")

# Find max salary per department for each region
df_employees.pivot_table(index=["Dept","Region"], values="Salary", observed=False, aggfunc="max")

# Find mean salary per department for each region
df_employees.pivot_table(index=["Dept"], columns=["Region"], values="Salary", observed=False, margins=True).round(2)
```

```python
# the unstack() method pivots the columns into a layer of indexes 
# It producess same result as a groupby
df_employees.pivot_table(index="Dept", columns="Region", values="Employee_ID", aggfunc="count", observed=False).unstack()
```

```python
# Proving that same result is obtained using groupby
df_employees.groupby(["Region","Dept"], observed=True)["Employee_ID"].count()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Generate synthetic data (e.g., heights of 1000 people)
# Mean = 170cm, Standard Deviation = 10cm
np.random.seed(42)
data = np.random.normal(loc=170, scale=10, size=1000)

# Create a figure with 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: The Histogram (Normalized)
# We use density=True so the total area of bars = 1
axs[0].hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
axs[0].set_title('1. The Histogram (Discrete Data)')
axs[0].set_xlabel('Value (e.g., Height)')
axs[0].set_ylabel('Density')

# Plot 2: Fitting the Curve (PDF)
# Calculate the PDF for the range of x
xmin, xmax = axs[0].get_xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 170, 10) # Normal distribution with our mean/std

axs[1].hist(data, bins=30, density=True, alpha=0.3, color='skyblue', edgecolor='grey')
axs[1].plot(x, p, 'k', linewidth=2, color='darkblue')
axs[1].set_title('2. Curve Fitted to Histogram')
axs[1].set_xlabel('Value (e.g., Height)')

# Plot 3: Calculating Area (Probability)
# Let's find probability of being between 160 and 180
axs[2].plot(x, p, 'k', linewidth=2, color='darkblue')
axs[2].set_title('3. Probability = Area Under Curve')
axs[2].set_xlabel('Value (e.g., Height)')

# Fill the area between 160 and 180
x_fill = np.linspace(160, 180, 100)
p_fill = norm.pdf(x_fill, 170, 10)
axs[2].fill_between(x_fill, p_fill, color='orange', alpha=0.6)
axs[2].text(170, 0.01, 'Area =\nProbability', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 175
sd = 7
x = np.linspace(mean - 4*sd, mean + 4*sd, 1000)
y = norm.pdf(x, mean, sd)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: CDF of Upper Bound (180)
ax1.plot(x, y, 'k')
fill_x = np.linspace(mean - 4*sd, 180, 1000)
fill_y = norm.pdf(fill_x, mean, sd)
ax1.fill_between(fill_x, fill_y, color='skyblue', alpha=0.8)
ax1.set_title(f'1. CDF(180cm)\nArea to left of 180 = 0.762')
ax1.axvline(180, color='blue', linestyle='--')

# Plot 2: CDF of Lower Bound (170)
ax2.plot(x, y, 'k')
fill_x = np.linspace(mean - 4*sd, 170, 1000)
fill_y = norm.pdf(fill_x, mean, sd)
ax2.fill_between(fill_x, fill_y, color='lightgrey', alpha=0.8)
ax2.set_title(f'2. CDF(170cm)\nArea to left of 170 = 0.238')
ax2.axvline(170, color='gray', linestyle='--')

# Plot 3: The Result (Difference)
ax3.plot(x, y, 'k')
# Fill the middle
mid_x = np.linspace(170, 180, 1000)
mid_y = norm.pdf(mid_x, mean, sd)
ax3.fill_between(mid_x, mid_y, color='lightgreen', alpha=0.8)
# Show the "removed" part faintly
ax3.fill_between(fill_x, fill_y, color='lightgrey', alpha=0.3)
ax3.set_title(f'3. The Difference\n0.762 - 0.238 = 0.524')
ax3.axvline(180, color='blue', linestyle='--')
ax3.axvline(170, color='gray', linestyle='--')

plt.tight_layout()
plt.show()
```
